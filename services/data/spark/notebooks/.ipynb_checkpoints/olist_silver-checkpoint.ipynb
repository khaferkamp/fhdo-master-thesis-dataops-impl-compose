## Table: customer

```python
%%sql
SELECT
    t.customer_id,
    t.customer_unique_id,
    t.customer_zip_code_prefix,
    t.customer_city,
    t.customer_state
FROM bronze.olist_customers t
WHERE 1=1
AND t.customer IS NOT NULL
AND t.customer_unique_id IS NOT NULL;
```

## Table: Geolocation

```python
%%sql
SELECT
    t.geolocation_zip_code_prefix,
    cast(t.geolocation_lat as float) as geolocation_lat,
    cast(t.geolocation_lng as float) as geolocation_lng,
    t.geolocation_city,
    t.geolocation_state
FROM bronze.olist_geolocation t
```

## Table: Order Items

```python
%%sql
SELECT
    t.order_id,
    cast(t.order_item_id as int) as order_item_id,
    t.product_id,
    t.seller_id,
    cast(t.shipping_limit_date as timestamp) as shipping_limit_date,
    cast(t.price as numeric(32,2)) price,
    cast(t.freight_value as numeric(32,2)) as freight_value
FROM bronze.olist_order_items t
```

## Table: Order Payments

```python
%%sql
SELECT
    t.order_id,
    cast(t.payment_sequential as int) as payment_sequential,
    t.payment_type,
    cast(t.payment_installments as int) as payment_installments,
    cast(t.payment_value as decimal(32,2)) as payment_value
FROM bronze.olist_order_payments t
```

## Table: Order Reviews

```python
%%sql
SELECT
    t.review_id,
    t.order_id,
    cast(t.review_score as int) as review_score,
    t.review_comment_title,
    t.review_comment_message,
    cast(t.review_creation_date as date) as review_creation_date,
    cast(t.review_answer_timestamp as timestamp) as review_answer_timestamp
FROM bronze.olist_order_reviews t
```

## Table: Orders

```python
%%sql
SELECT
    t.order_id,
    t.customer_id,
    t.order_status,
    cast(t.order_purchase_timestamp as timestamp) as order_purchase_timestamp,
    cast(t.order_approved_at as timestamp) as order_approved_at,
    cast(t.order_delivered_carrier_date as date) as order_delivered_carrier_date,
    cast(t.order_delivered_customer_date as date) as order_delivered_customer_date,
    cast(t.order_estimated_delivery_date as date) as order_estimated_delivery_date
FROM bronze.olist_orders t
```

## Table: Products

```python
%%sql
SELECT
    t.product_id,
    t.product_category_name,
    t.product_name_lenght,
    t.product_description_lenght,
    t.product_photos_qty,
    t.product_weight_g,
    t.product_length_cm,
    t.product_height_cm,
    t.product_width_cm
FROM temp_products t
```

## Table: Sellers

```python
sellers_df = (spark
     .read
     .option("header", "true")
     .csv("/var/lib/ngods/stage/olist/olist_sellers_dataset.csv"))
```

```python
sellers_df.createOrReplaceTempView("temp_sellers")
```

```python
sellers_df.describe()
```

```python
%%sql
CREATE TABLE IF NOT EXISTS bronze.olist_sellers (
    seller_id string,
    seller_zip_code_prefix string,
    seller_city string,
    seller_state string
)
USING iceberg;
```

```python
%%sql
MERGE INTO bronze.olist_sellers t
USING (
    SELECT
    seller_id,
    seller_zip_code_prefix,
    seller_city,
    seller_state
    FROM temp_sellers) s
ON t.seller_id = s.seller_id
WHEN NOT MATCHED THEN INSERT *
```

## Table: Product Category Name Translation

```python
pcnt_df = (spark
     .read
     .option("header", "true")
     .csv("/var/lib/ngods/stage/olist/product_category_name_translation.csv"))
```

```python
pcnt_df.createOrReplaceTempView("temp_pcnt")
```

```python
pcnt_df.describe()
```

```python
%%sql
CREATE TABLE IF NOT EXISTS bronze.olist_product_category_name_translation (
    product_category_name string,
    product_category_name_english string
)
USING iceberg;
```

```python
%%sql
MERGE INTO bronze.olist_product_category_name_translation t
USING (
    SELECT
    product_category_name,
    product_category_name_english
    FROM temp_pcnt) s
ON t.product_category_name = s.product_category_name
WHEN NOT MATCHED THEN INSERT *
```
